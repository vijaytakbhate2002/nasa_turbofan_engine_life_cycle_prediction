```python
import numpy as np 
import pandas as pd 
pd.set_option('display.max_columns', None)
import os 
```

```python
ROOT_PATH = "/kaggle/input/nasa-cmaps/CMaps"
METADATA = ['readme.txt', 'x.txt', 'Damage Propagation Modeling.pdf']
DATA_TXT = []

for file in os.listdir(ROOT_PATH):
    if file not in METADATA:
        DATA_TXT.append(file)
DATA_TXT
```

## Merge Train and Test Data into dataframe

```python
train_df = pd.DataFrame()
test_df = pd.DataFrame()
rul_df = pd.DataFrame()

for file in DATA_TXT:
    if 'train' in file:
        df = pd.read_csv(os.path.join(ROOT_PATH, file), sep=' ', header=None)
        df['file_name'] = file
        train_df = pd.concat([train_df, df], axis=0)
    elif 'test' in file:
        df = pd.read_csv(os.path.join(ROOT_PATH, file), sep=' ', header=None)
        # print(df.shape)
        df['file_name'] = file
        test_df = pd.concat([test_df, df], axis=0)
    elif 'RUL' in file:
        df =  pd.read_csv(os.path.join(ROOT_PATH, file), sep=' ', header=None)
        rul_df['file_name'] = file
        rul_df = pd.concat([rul_df, df], axis=0)
        
train_df.head()
```

## Making data suitable for training our model

```python
train_df['engine_id'] = train_df['file_name'] + '_' + train_df[0].astype(str)
test_df['engine_id'] = test_df['file_name'] + '_' + test_df[0].astype(str)
train_df['engine_id'].value_counts()
```

**Add RUL for each row in training and testing data**

```python
# train_df[train_df['engine_id'] == 'train_FD004.txt_118']
train_rul = []
for engine_id, cycles in dict(train_df['engine_id'].value_counts()).items():
    temp = list(range(1, cycles+1))
    temp.reverse()
    train_rul += temp

train_df['RUL'] = train_rul
train_df.head()
```

```python
test_rul = []
for rul, (engine_id, cycles) in zip(rul_df[0], dict(test_df['engine_id'].value_counts()).items()):
    temp = list(range(1, cycles + 1 + int(rul)))
    temp.reverse()
    test_rul += temp[:cycles]

test_df['RUL'] = test_rul
test_df.head()
```

**Divide each engine data while keeping 19 sequences in a single chunk**

```python
from sklearn.preprocessing import LabelEncoder

train_df = train_df.drop(['file_name'], axis='columns')
test_df = test_df.drop(['file_name'], axis='columns')

encoder = LabelEncoder()
train_df['engine_id'] = encoder.fit_transform(train_df['engine_id'])
test_df['engine_id'] = encoder.fit_transform(test_df['engine_id'])

train_df.head()
```

## Normalizing data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_df_norm = scaler.fit_transform(train_df.drop(['engine_id', 'RUL', 0, 1], axis='columns'))
train_df_norm = pd.DataFrame(train_df_norm, columns = scaler.get_feature_names_out())
train_df_norm['Cycle_Number'] = train_df[1].values
train_df_norm['engine_id'] = train_df['engine_id'].values
train_df_norm['RUL'] = train_df['RUL'].values

test_df_norm = scaler.transform(test_df.drop(['engine_id', 'RUL', 0, 1], axis='columns'))
test_df_norm = pd.DataFrame(test_df_norm, columns = scaler.get_feature_names_out())
test_df_norm['Cycle_Number'] = test_df[1].values
test_df_norm['engine_id'] = test_df['engine_id'].values
test_df_norm['RUL'] = test_df['RUL'].values

train_df_norm.head()
```

```python
import sys

train_X, test_X = [], []
train_y, test_y = [], []


for i, (engine, cycles) in enumerate(dict(train_df_norm['engine_id'].value_counts(sort=True)).items()):
    sys.stdout.flush()
    print('train df: applying window for engine number = ', i + 1, " out of ", len(train_df['engine_id'].unique()), end="\r")
    train_temp = train_df_norm.drop(['RUL'], axis='columns').astype('float16')
    for i in range(cycles-10):
        
        train_x_temp = np.array(train_temp.iloc[i:i+10])
        train_y_temp = train_df_norm['RUL'].iloc[i + 10]

        train_X.append(train_x_temp)
        train_y.append(train_y_temp)


for i, (engine, cycles) in enumerate(dict(test_df_norm['engine_id'].value_counts(sort=True)).items()):
    sys.stdout.flush()
    print('test_df: applying window for engine number = ', i + 1, " out of ", len(test_df['engine_id'].unique()), end="\r")
    test_temp = test_df_norm.drop(['RUL'], axis='columns').astype('float16')
    for i in range(cycles-10):
        
        test_x_temp = np.array(test_temp.iloc[i:i+10])
        test_y_temp = test_df_norm['RUL'].iloc[i + 10]

        test_X.append(test_x_temp)
        test_y.append(test_y_temp)


train_X = np.array(train_X)
test_X = np.array(test_X)

train_y = np.array(train_y)
test_y = np.array(test_y)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, LSTM
```

```python

# Get input shape
timesteps = train_X.shape[1]   # e.g. window length
features = train_X.shape[2]    # number of features per timestep

model = Sequential([
    # --- CNN part ---
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(timesteps, features)),
    MaxPooling1D(pool_size=2),
    # Dropout(0.3),
    
    # --- LSTM part ---
    LSTM(64, return_sequences=False),
    # Dropout(0.3),
    
    # --- Fully connected ---
    Dense(64, activation='relu'),
    Dense(1)   # output layer → RUL (regression)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# Fit model
history = model.fit(train_X, train_y, 
                    validation_split=0.3,
                    epochs=20, 
                    batch_size=64,
                    verbose=1)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

class Evaluate:
    
    def __init__(self):
        pass
        
    def accuracy(self, y_true, y_pred):
        error = 0
        max_y = max(y_true)
        min_y = min(y_true)
        
        for y_t, y_p in zip(y_true, y_pred):
            error += abs(y_t - y_p)
        
        avg_error = error / len(y_true)
        return 1 - (avg_error / (max_y - min_y))

    def evaluate(self, y_true, y_pred):
        y_true = list(y_true)
        y_pred = list(y_pred)
        accuracy = self.accuracy(y_true, y_pred)
        mse = mean_squared_error(y_true, y_pred)
        mae = mean_absolute_error(y_true, y_pred)
        return {"ACCURACY": accuracy, 'MSE': mse, 'MAE': mae}
```

```python
y_pred_train = model.predict(train_X).reshape(1, -1)[0]
print("Evaluation on training data: ")
evaluator = Evaluate()
evaluator.evaluate(train_y, y_pred_train)
```

```python
y_pred_test = model.predict(test_X).reshape(1, -1)[0]
print("Evaluation on training data: ")
evaluator = Evaluate()
evaluator.evaluate(test_y, y_pred_test)
```
